```python
from bs4 import BeautifulSoup
import requests
from lxml import etree


url = "https://www.immoscoop.be/te-koop/2960-sint-job-in-t-goor/298009"
r = requests.get(url)
print(url, r.status_code)
soup = BeautifulSoup(r.content, "html.parser")
soup
```

```python
summary = {}

# 1) LOCALITY
locality = ""
for elem in soup.find("address", attrs={"class": "not-italic text-sm"}):
    locality += elem.text
summary["locality"] = locality

# 2) Type of residence *house appartment
type_residence = ''
dom = etree.HTML(str(soup))
type = dom.xpath('/html/body/div[1]/div/div/main/div[2]/div/div/div[2]/div/div[1]/div[2]')
type_residence += type[0].text
summary["type_residence"] = type_residence


# 3) Subtype of property (Bungalow, Chalet, Mansion, ...)
subtype_residence = ''
subtype = dom.xpath('/html/body/div[1]/div/div/main/div[2]/div/div/div[2]/div/div[2]/div[2]')
subtype_residence += subtype[0].text
summary["subtype_residence"] = subtype_residence

# 4) Price
price = dom.xpath('/html/body/div[1]/div/div/main/div[1]/div[2]/div[2]/div')
price = price[0].text
price = price.split('€\xa0')[1].split('.')
price = int(''.join(price))
summary["price"] = price

# 5) Type of sale (Exclusion of life sales)
# ?????

# 6) Number of rooms
number_of_rooms = dom.xpath('/html/body/div[1]/div/div/main/div[2]/div/div/div[5]/div/div[1]/div[2]')
number_of_rooms = int(number_of_rooms[0].text)
summary["number_of_rooms"] = number_of_rooms

# 7) Living Area

# 8) Fully equipped kitchen (Yes/No)

# 9) Furnished (Yes/No)

# 10) Open fire (Yes/No)

# 11) Terrace (Yes/No)
 #  If yes: Area

# 12) Garden (Yes/No)
 #  If yes: Area

# 13) Surface of the land

# 14) Surface area of the plot of land

# 15) Number of facades

# 16) Swimming pool (Yes/No)

# 17) State of the building (New, to be renovated, ...)
```
